| **1**                   | **Apprentissage Automatique Traditionnel**     | **Apprentissage Profond**                       |
|------------------------------|-----------------------------------------------|------------------------------------------------|
| **Ingénierie des fonctionnalités** | Nécessite une extraction manuelle des caractéristiques par des experts. | Automatisée, le modèle apprend directement à partir des données brutes. |
| **Informatique**             | Peut fonctionner sur des CPU classiques.     | Nécessite des GPU/TPU puissants pour l'entraînement. |
| **Évolutivité**              | Moins évolutif sur de grands ensembles de données. | Très évolutif avec des grandes quantités de données. |
| **Découverte de modèles**    | Basé sur des hypothèses et des règles prédéfinies. | Apprend des représentations complexes sans intervention humaine. |
| **Exigences informatiques**  | Faibles à modérées, fonctionne avec des ressources limitées. | Élevées, demande beaucoup de mémoire et de puissance de calcul. |


### 2 Exemples de problèmes adaptés à chaque approche :

| **Type d'apprentissage**              | **Problème concret** |
|--------------------------------------|----------------------|
| **Apprentissage automatique traditionnel** | **Détection de fraude bancaire** : Les modèles comme la régression logistique ou les arbres de décision sont efficaces pour détecter les transactions suspectes en analysant des caractéristiques tabulaires (montant, lieu, fréquence des transactions). |
| **Apprentissage profond**            | **Reconnaissance d’images** : Un CNN (réseau de neurones convolutif) est idéal pour identifier des objets dans des images (ex : détection de panneaux de signalisation pour les voitures autonomes). |

L'**apprentissage automatique traditionnel** est souvent préféré lorsque les données sont structurées et que l’interprétabilité est importante.  
L'**apprentissage profond** est plus puissant pour traiter des données non structurées comme les images, le texte et l’audio.


## 3
L'apprentissage en profondeur est idéal pour les données non structurées (images, texte, audio) car il apprend automatiquement des patrons complexes sans ingénierie des caractéristiques. Grâce à ses couches cachées, il extrait des représentations utiles, rendant les CNN efficaces pour les images et les Transformers pour le texte. Cependant, il nécessite beaucoup de donnéeset de puissance de calcul.

```python
#exp2
import matplotlib.pyplot as plt
import networkx as nx

# Create a simple artificial neural network
G = nx.DiGraph()

# Add layers
input_layer = ['I1', 'I2', 'I3']
hidden_layer = ['H1', 'H2', 'H3', 'H4']
output_layer = ['O1', 'O2']

# Add nodes for input layer, hidden layer, and output layer
G.add_nodes_from(input_layer, layer='input')
G.add_nodes_from(hidden_layer, layer='hidden')
G.add_nodes_from(output_layer, layer='output')

# Add edges between layers (representing weights)
for i in input_layer:
    for h in hidden_layer:
        G.add_edge(i, h)

for h in hidden_layer:
    for o in output_layer:
        G.add_edge(h, o)

# Set the positions of the nodes
pos = {}

# Set positions for input, hidden, and output layers
for i, node in enumerate(input_layer):
    pos[node] = (0, i)
for i, node in enumerate(hidden_layer):
    pos[node] = (1, i)
for i, node in enumerate(output_layer):
    pos[node] = (2, i)

# Draw the network
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='skyblue', font_size=10, font_weight='bold', arrows=True)

# Add labels for layers and components
plt.text(-0.1, 1.5, 'Input Layer', fontsize=12, ha='center')
plt.text(1.1, 1.5, 'Hidden Layer', fontsize=12, ha='center')
plt.text(2.1, 1.5, 'Output Layer', fontsize=12, ha='center')

# Annotating weights, activation functions, and bias
for i in input_layer:
    for h in hidden_layer:
        plt.text(0.1, hidden_layer.index(h), 'Weight', fontsize=9, color='green')
for h in hidden_layer:
    for o in output_layer:
        plt.text(1.1, output_layer.index(o), 'Weight', fontsize=9, color='green')

# Biases in hidden and output layers
for h in hidden_layer:
    plt.text(1.1, hidden_layer.index(h) + 0.2, 'Bias', fontsize=9, color='orange')
for o in output_layer:
    plt.text(2.1, output_layer.index(o) + 0.2, 'Bias', fontsize=9, color='orange')

# Activation functions
for h in hidden_layer:
    plt.text(1.1, hidden_layer.index(h) - 0.3, 'ReLU Activation', fontsize=9, color='red')
for o in output_layer:
    plt.text(2.1, output_layer.index(o) - 0.3, 'Softmax Activation', fontsize=9, color='red')

# Set title
plt.title("Simple Artificial Neural Network (ANN) with Components")

plt.show()
```

Dans un réseau de neurones artificiels, l'information circule de manière séquentielle à travers différentes couches. Les entrées du réseau (couche d'entrée) sont transmises aux neurones de la couche cachée via des connexions pondérées, où chaque poids influence la force de la connexion. Les neurones de la couche cachée appliquent une fonction d'activation (comme ReLU) pour transformer les données avant de les envoyer à la couche de sortie. Enfin, la couche de sortie produit les résultats finaux, souvent après avoir appliqué une fonction d'activation comme Softmax, qui génère des probabilités dans le cas d'une classification. Ce processus de propagation de l'information permet au réseau de faire des prédictions ou de prendre des décisions

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixer la graine pour la reproductibilité
np.random.seed(0)

# Générer les valeurs de x
x = np.linspace(-1, 1, 20)   # 20 points allant de -1 à 1

# Calculer les valeurs de y avec du bruit ajouté
y = -x**2 + np.random.normal(0, 0.05, len(x))

# Afficher les résultats
plt.scatter(x, y, color='blue', label='Données avec bruit')
plt.plot(x, -x**2, color='red', label='y = -x^2 (fonction réelle)', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ensemble de points générés avec bruit')
plt.legend()
plt.show()
```

```python
x_train = x[:12]  # 12 premiers points
y_train = y[:12]  # 12 premiers points
x_test = x[12:]   # 8 derniers points
y_test = y[12:]   # 8 derniers points

plt.scatter(x_train, y_train, color='blue', label='Données d\'entraînement')
plt.scatter(x_test, y_test, color='green', label='Données de test')
plt.plot(x, -x**2, color='red', label='y = -x^2 (fonction réelle)', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ensemble de données générées avec bruit (Entraînement et Test)')
plt.legend()
plt.show()
```
