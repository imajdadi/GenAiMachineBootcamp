```python
# Step 1: Import nltk
import nltk
# Step 2: Download punkt and punkt_tab
nltk.download('punkt')
nltk.download('punkt_tab')  # Attempt to download punkt_tab as suggested
# Step 3: Define the text
text = "Why, sometimes I've believed as many as 6 impossible things before breakfast?"
# Step 4: Tokenize the text using word_tokenize()
from nltk.tokenize import word_tokenize
word_tokens = word_tokenize(text)
print("Word Tokens:", word_tokens)
# Step 5: Add another sentence after the question mark
text += " It's a curious thought, isn't it?"
# Step 6: Use sent_tokenize() to print each sentence
from nltk.tokenize import sent_tokenize
sentence_tokens = sent_tokenize(text)
print("\nSentence Tokens:")
for sentence in sentence_tokens:
    print(sentence)
```

```python
from nltk.corpus import stopwords
import re
doc = '''Linguistics and Natural Language Processing (NLP) are closely linked. \n Linguistics is the scientific study of language, encompassing its structure, meaning, and context. \n It provides foundational knowledge about language syntax, semantics, pragmatics, and phonetics. \n NLP applies this linguistic knowledge in computational algorithms to enable computers to understand, interpret, and generate human language. By leveraging linguistic principles, NLP seeks to bridge the gap between human communication and computer understanding, enabling tasks like translation, sentiment analysis, and voice recognition.'''
clean_doc = re.sub(r'[^\w\s]', '', doc)  # Supprimer la ponctuation
clean_doc = re.sub(r'\n', ' ', clean_doc)
tokens = word_tokenize(clean_doc)
# Liste des stopwords en anglais
stop_words = set(stopwords.words('english'))
# Filtrer les tokens pour exclure les stopwords
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

# Afficher les tokens d'origine et les tokens filtrés
print("Jetons d'origine:", tokens)
print("Jetons filtrés:", filtered_tokens)
```

```python
#Correction 2 :
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
# Télécharger les ressources nécessaires
nltk.download('punkt')
nltk.download('stopwords')
# Texte à traiter
doc = '''Linguistics and Natural Language Processing (NLP) are closely linked. \n Linguistics is the scientific study of language, encompassing its structure, meaning, and context. \n It provides foundational knowledge about language syntax, semantics, pragmatics, and phonetics. \n NLP applies this linguistic knowledge in computational algorithms to enable computers to understand, interpret, and generate human language. By leveraging linguistic principles, NLP seeks to bridge the gap between human communication and computer understanding, enabling tasks like translation, sentiment analysis, and voice recognition.'''
# Nettoyer le texte en supprimant les caractères non alphabétiques
doc_cleaned = re.sub(r'\W+', ' ', doc)
# Tokeniser le texte
tokens = word_tokenize(doc_cleaned)
# Filtrer les mots vides
filtered_words = [word for word in tokens if word.lower() not in stopwords.words('english')]
# Afficher les tokens originaux et filtrés
print("Original Tokens:", tokens)
print("Filtered Tokens:", filtered_words)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import spacy
# Télécharger les ressources nécessaires
nltk.download('punkt')
nltk.download('stopwords')
# Texte à traiter
doc_text = '''Charles Lutwidge Dodgson, better known by his pen name Lewis Carroll, was an English author, poet, mathematician and photographer. His most notable works are Alice's Adventures in Wonderland (1865) and its sequel Through the Looking-Glass (1871).'''
# Tokenisation avec NLTK
tokens = word_tokenize(doc_text)
# Initialiser le stemmer
stemmer = PorterStemmer()
# Appliquer le stemming aux tokens
stemmed = [stemmer.stem(word) for word in tokens]
print('Stemmed:', stemmed)
# Charger le modèle de langue anglaise de spaCy
nlp = spacy.load('en_core_web_sm')
# Traiter le texte avec spaCy
doc = nlp(doc_text)
# Filtrer les mots vides et les caractères non alphabétiques
stop_words = set(stopwords.words('english'))
filtered_doc = [token for token in doc if token.text.lower() not in stop_words and token.is_alpha]
# Appliquer la lemmatisation
lemmatized = [token.lemma_ for token in filtered_doc]
print('Lemmatized:', lemmatized)
```

```python
from nltk import pos_tag, ne_chunk
from nltk.tokenize import word_tokenize
# Télécharger les ressources nécessaires
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker_tab', force=True)
nltk.download('words')
# Texte à traiter
doc_text = '''Charles Lutwidge Dodgson, better known by his pen name Lewis Carroll, was an English author, poet, mathematician and photographer. His most notable works are Alice's Adventures in Wonderland (1865) and its sequel Through the Looking-Glass (1871).'''
# Tokenisation avec NLTK
tokens = word_tokenize(doc_text)
# Appliquer le POS tagging
tagged = pos_tag(tokens)
# Appliquer la reconnaissance d'entités nommées (NER)
entities = ne_chunk(tagged)
# Afficher les entités reconnues
print(entities)
```
