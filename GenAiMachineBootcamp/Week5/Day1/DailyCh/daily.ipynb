```python
import requests
import re
from bs4 import BeautifulSoup

def load_texts(urls):
    corpus = []
    
    for url in urls:
        try:
            response = requests.get(url, timeout=10)
            response.raise_for_status()  # Vérifie si la requête a réussi
            
            soup = BeautifulSoup(response.text, 'html.parser')
            text = soup.get_text(separator=' ')  # Extrait le texte de la page
            
            # Nettoyage du texte : suppression des caractères spéciaux et non-mots
            cleaned_text = re.sub(r'[^\w\s]', '', text)  # Supprime tout sauf les mots et espaces
            cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()  # Supprime les espaces multiples
            
            corpus.append(cleaned_text)
        except requests.RequestException as e:
            print(f"Erreur lors de l'accès à {url}: {e}")
    
    return corpus

urls = [
    "https://www.gutenberg.org/cache/epub/29042/pg29042.txt",
    "https://www.gutenberg.org/cache/epub/12/pg12.txt",
    "https://www.gutenberg.org/cache/epub/11/pg11.txt"
]
corpus = load_texts(urls)
print(corpus)

            
```

```python
import requests
from bs4 import BeautifulSoup

def extract_relevant_text_from_url(url):
    # Récupérer le contenu HTML de la page
    response = requests.get(url)
    
    # Vérifier si la requête a réussi (status code 200)
    if response.status_code != 200:
        return "Erreur lors du chargement de la page"
    
    # Parser le contenu HTML
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extraire tout le texte visible de la page (en supprimant les balises HTML)
    text = soup.get_text()
    
    # Chercher la position de 'START' et '*** END' dans le texte
    start_index = text.find('START')
    end_index = text.find('*** END')
    
    # Extraire la partie pertinente entre 'START' et '*** END'
    if start_index != -1 and end_index != -1:
        relevant_text = text[start_index + len('START'):end_index].strip()
    else:
        relevant_text = text.strip()
    
    # Limiter à 200 premiers caractères
    return relevant_text[:200]

# Liste des URLs
urls = [
    "https://www.gutenberg.org/cache/epub/29042/pg29042.txt",
    "https://www.gutenberg.org/cache/epub/12/pg12.txt",
    "https://www.gutenberg.org/cache/epub/11/pg11.txt"
]

# Appeler la fonction pour chaque URL et afficher les 200 premiers caractères
for url in urls:
    print(extract_relevant_text_from_url(url))
```

```python

import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Télécharger les ressources nécessaires de NLTK pour les mots vides (stopwords)
nltk.download('punkt')
nltk.download('stopwords')

# Fonction pour récupérer le texte à partir de l'URL
def get_text_from_url(url):
    response = requests.get(url)
    if response.status_code != 200:
        return None
    soup = BeautifulSoup(response.text, 'html.parser')
    text = soup.get_text()
    return text

# Fonction pour tokeniser le texte
def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens

# Fonction pour supprimer les mots vides
def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))  # Liste des mots vides en anglais
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return filtered_tokens

# Fonction pour vérifier la suppression des mots vides
def check_stopwords_removed(original_tokens, filtered_tokens):
    original_count = len(original_tokens)
    filtered_count = len(filtered_tokens)
    print(f"Nombre de tokens originaux : {original_count}")
    print(f"Nombre de tokens après suppression des mots vides : {filtered_count}")
    print(f"Mots vides supprimés : {original_count - filtered_count}")
    
    # Recherche de certains mots vides spécifiques
    stopwords_to_check = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves']
    for word in stopwords_to_check:
        print(f"{word}: {filtered_tokens.count(word)}")

urls = [
    "https://www.gutenberg.org/cache/epub/29042/pg29042.txt",
    "https://www.gutenberg.org/cache/epub/12/pg12.txt",
    "https://www.gutenberg.org/cache/epub/11/pg11.txt"
]

# Appliquer la fonction à chaque URL
for url in urls:
    print(f"Traitement de {url}...\n")
    text = get_text_from_url(url)
    if text:
        tokens = tokenize_text(text)
        filtered_tokens = remove_stopwords(tokens)
        check_stopwords_removed(tokens, filtered_tokens)
        print("\n" + "-"*50 + "\n")
    else:
        print(f"Erreur lors du chargement du texte depuis {url}\n")

```

```python
import requests
import nltk
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# Télécharge les ressources nécessaires pour nltk
nltk.download('punkt')

def extract_relevant_text_from_url(urls):
    # Récupérer le contenu HTML de la page
    response = requests.get(urls)
    
    # Vérifier si la requête a réussi (status code 200)
    if response.status_code != 200:
        return "Erreur lors du chargement de la page"
    
    # Parser le contenu HTML
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extraire tout le texte visible de la page (en supprimant les balises HTML)
    text = soup.get_text()
    
    return text

urls = [
    "https://www.gutenberg.org/cache/epub/29042/pg29042.txt",
    "https://www.gutenberg.org/cache/epub/12/pg12.txt",
    "https://www.gutenberg.org/cache/epub/11/pg11.txt"
]

# Extraire le texte de l'URL
text = extract_relevant_text_from_url(urls)

# Tokenisation du texte
tokens = word_tokenize(text)

# Appliquer le stemming avec PorterStemmer
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in tokens]

# Afficher les 50 premiers jetons stemmés
print(stemmed_tokens[:50])
```
