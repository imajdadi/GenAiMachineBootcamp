```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data= np.random.uniform(-5, 35, (12, 10))
df= pd.DataFrame(
    data, 
            index = ["Jan","Fev","Mar","Avr","Mai","Juin","Jui","Aout","Sept","Octo","Nove","Dec" ]
            ,columns =["Paris", "Nice", "Chambery", "Saint-tropez", "Lyon", "Annecy", "Etretat" , "Brest", "toulon", "Nantes" ]
            )
print(df)
```

```python
# Trouver la moyenne
moyenne_annuelle_ville = df.mean(axis=0)

print("Température moyenne annuelle par ville :\n",moyenne_annuelle_ville)
```

```python
# Calcul de la température moyenne par ville (par colonne)
moyenne_par_ville = df.mean(axis=0)

# Ville avec la température moyenne la plus élevée
ville_max_temp = moyenne_par_ville.idxmax()

# Ville avec la température moyenne la plus basse
ville_min_temp = moyenne_par_ville.idxmin()
# Afficher le meilleur client
print("La ville avec la température moyenne la plus élevée est", ville_max_temp)
print("La ville avec la température moyenne la plus basse est ", ville_min_temp)
```

```python
# Visualisation
df.plot(kind='line', figsize=(12, 7))

plt.title("Températures Mensuelles par Ville")
plt.xlabel("Mois")
plt.ylabel("Température (°C)")
plt.legend(title="Villes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

L'analyse des températures mensuelles moyennes sur 12 mois dans 10 villes françaises montre qu'Annecy a enregistré les températures les plus élevées, tandis qu'Etretat a affiché les plus basses. Les villes situées dans les régions montagneuses, comme Annecy, connaissent des températures plus chaudes, tandis que celles près du littoral, comme Etretat, restent plus fraîches. Les données révèlent également des variations saisonnières marquées, avec des températures les plus basses en hiver et les plus élevées en été, ce qui peut influencer des décisions liées au climat ou au tourisme.
