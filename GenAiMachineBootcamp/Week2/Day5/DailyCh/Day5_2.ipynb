```python
import pandas as pd
# Charger le fichier Excel
df = pd.read_excel("US Superstore data.xls")
# Afficher les premières lignes
print(df.head())
```

```python
print("Types de données dans chaque colonne:")
print(df.dtypes)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Grouper les ventes par État et trier par ordre décroissant
product_ventes = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

# Ajuster la taille de la figure
plt.figure(figsize=(15, 10))

# Modifier la taille de la police pour améliorer la lisibilité
plt.rcParams.update({'font.size': 7})

# Créer un graphique à barres avec seaborn
sns.barplot(x=product_ventes.index, y=product_ventes.values, palette="Blues_r")

# Rotation des étiquettes sur l'axe X pour éviter le chevauchement
plt.xticks(rotation=45)

# Ajouter des étiquettes aux axes et un titre
plt.xlabel("Produits")
plt.ylabel("les ventes")
plt.title("Top des 10 Prdoduits avec le plus de ventes")

# Afficher le graphique
plt.show()
# Afficher les 5 États avec le plus de ventes
print(product_ventes.head())
```

```python
#Créez un nuage de points pour analyser la relation entre le bénéfice et la remise.
sns.scatterplot(data=df, x='Discount', y='Profit')

# Ajout d'un titre et des labels des axes
plt.title('la relation entre le bénéfice et la remise')
plt.xlabel('les remises')
plt.ylabel('benefices')


# Affichage du graphique
plt.show()
```

Matplotlib permet un contrôle détaillé sur les graphiques, mais demande plus de code et d'ajustements manuels. Seaborn, en revanche, simplifie la création de visualisations grâce à des styles intégrés et des fonctionnalités statistiques. Pour des graphiques personnalisés et complexes, Matplotlib est plus adapté, tandis que Seaborn est idéal pour des visualisations rapides et esthétiques avec moins d’effort.
