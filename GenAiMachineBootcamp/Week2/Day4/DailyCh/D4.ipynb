```python
import pandas as pd
df = pd.read_csv('2019.csv')
print(df.head())
print("Types de données dans chaque colonne:")
print(df.dtypes)
print(df.info()) 
```

```python
# Vérifier les valeurs manquantes par colonne
missing_values = df.isnull().sum()
# Afficher uniquement les colonnes qui ont des valeurs manquantes
missing_columns = missing_values[missing_values > 0]
print("Colonnes avec des valeurs manquantes :\n", missing_columns)

# Suppression des colonnes où toutes les valeurs sont manquantes
#df_cleaned = df.dropna(axis=1, how='all')
#DataFrame ne contient aucune valeur manquante

 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Création du nuage de points
sns.scatterplot(data=df, x='Social support', y='Score')

# Ajout d'un titre et des labels des axes
plt.title('la relation entre « soutien social » et « score »')
plt.xlabel('soutien social')
plt.ylabel('score')


# Affichage du graphique
plt.show()
```

```python
# Sélection de la colonne par son nom
colonne = 'Country or region'
contenu_colonne = df[colonne]

# Affichage du contenu de la colonne
print(contenu_colonne)
g= sns.jointplot(x='GDP per capita', y='Healthy life expectancy', data=df, kind='scatter', hue='Country or region', palette='deep')
# Affichage du graphique
# Récupérer la légende
legend = g.ax_joint.get_legend()

# Ajuster la taille de la police de la légende
for text in legend.get_texts():
    text.set_fontsize(10)  # Ajustez la taille selon vos besoins

# Positionner la légende en dehors du graphique
legend.set_bbox_to_anchor((1.75, 1))
legend.set_title('Régions et pays')

# Afficher le graphique
plt.show()
```
