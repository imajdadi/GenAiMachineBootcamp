```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
data = sns.load_dataset('tips')
sns.histplot(data['total_bill'], bins=15, color='blue')
plt.title('Histogram customer spending')
plt.show()
```

```python
iris = sns.load_dataset('iris')
g = sns.PairGrid(iris, hue='species') #markers = ['o', 's', 'D'])
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap='Blues')
g.map_diag(sns.histplot, kde=True, color='teal')
g.add_legend()
plt.show()
```

```python
data = sns.load_dataset('titanic')
g = sns.FacetGrid(data, col='sex', row='class', margin_titles=True)
g.map(sns.histplot,'age', color='blue')#gmap fonction 

plt.show()
```
