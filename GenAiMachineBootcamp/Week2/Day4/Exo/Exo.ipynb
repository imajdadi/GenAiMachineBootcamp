```python
#Exo5
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des données échantillons
data =  {
    'Anxiety': ["oui", "Non", "oui", "Non", "oui"],
    'Sex': ["femelle", "femelle", "femelle", "mal", "mal"]
}
df = pd.DataFrame(data)



```

```python
# Calcul du nombre total d'étudiants par sexe
total_par_sexe = df.groupby("Sex").size()

# Calcul du nombre d'étudiants souffrant d'anxiété par sexe 
anxieux_par_sexe = df[df["Anxiety"].str.lower() == "oui"].groupby("Sex").size()

# Calcul de la proportion
proportions = (anxieux_par_sexe / total_par_sexe).reset_index(name="Proportion")

# Création du graphique à barres avec Seaborn
sns.set_theme(style="whitegrid")
plt.figure(figsize=(6,4))
ax = sns.barplot(x="Sex", y="Proportion", data=proportions, palette="viridis")

# Personnalisation du graphique
plt.title("Proportion d'étudiants souffrant d'anxiété selon le sexe")
plt.xlabel("Sexe")
plt.ylabel("Proportion d'anxiété")

# Affichage du tracé
plt.show()
```

```python
#Exo6
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des données échantillons
data =  {
    'age': [12 , 15 , 17 , 10 , 9],
    'attaques': ["non", "oui", "oui", "non", "oui"]
}
df = pd.DataFrame(data)

```

```python
df["Attaques_numeriques"] = df["attaques"].str.lower().map({"oui": 1, "non": 0})
#g = sns.PairGrid(df, vars=["age", "Attaques_numeriques"] )
#g = sns.PairGrid(df, col='Age', row='Attaques') #markers = ['o', 's', 'D'])
# Création du nuage de points
sns.scatterplot(data=df, x='age', y='Attaques_numeriques')

# Ajout d'un titre et des labels des axes
plt.title('Relation entre l\'Âge et les Attaques de Panique')
plt.xlabel('Âge')
plt.ylabel('Attaques de Panique (1 = Oui, 0 = Non)')


# Affichage du graphique
plt.show()
```
