```python
#Exo1
import matplotlib.pyplot as plt

x = ["Lundi","Mardi" , "Mercredi", "Jeudi", "Vendredi" , "Samedi" , "Dimanche"]
y =[72, 74, 76, 80, 82, 78, 75]

plt.title("Graphique les variations de température ")
plt.xlabel("les jours de la semaine")
plt.ylabel("Température (°F)")
plt.plot(x, y)
plt.show()
```

```python
#Exo2
categories = ['Janvier', 'Fevrier', 'Mars', 'Avril', 'Mai']
values = [5000, 5500, 6200, 7000, 7500]

plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'], edgecolor='black' )#hatch='/'
plt.title('Les Donnees de ventes mensuelles')
plt.xlabel('Mois')
plt.ylabel('Montant des ventes ($)')
plt.show()
```

```python
#Exo4
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load sample data
data = {'CGPA': [2.5, 3.0, 3.2, 3.5, 3.7, 3.8, 4.0, 3.1, 2.8, 3.6]}
df = pd.DataFrame(data)
#sns.histplot(data['total_bill'], bins=15, color='blue')
sns.histplot(df['CGPA'], bins=20, color='red')
plt.title('Histogram of CGPA')
plt.show()
```
