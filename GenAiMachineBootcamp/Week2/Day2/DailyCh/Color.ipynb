```python
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt

# Load an image using PIL
image_path = 'Flower Color Images Dataset/Flower Color Images Dataset/flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
plt.axis("off")
```

```python
from scipy.ndimage import rotate
def rotate_image_90_degrees(image):
    return rotate(image, angle=90, reshape=False, mode='nearest')
    
# Appliquer la rotation
rotated_image = rotate_image_90_degrees(original_image)

# Afficher l'image pivotée
plt.imshow(rotated_image)
plt.axis("off")
plt.show()
```

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

mirrored_image = ImageOps.mirror(original_image)

# Appliquer le retournement vertical (haut-bas)
flipped_image = ImageOps.flip(original_image)

plt.subplot(1, 2, 1)
plt.imshow(mirrored_image)

plt.subplot(1, 2, 2)
plt.imshow(flipped_image)


```
